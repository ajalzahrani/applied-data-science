Pandas functions for data analysis operations are gouped under exhaustive categories such as:
* General functions
* Input / output functions
* Series functions
* DataFrame functions
* Resampling
* Plotting
* Statistical functions
* Data and Timedelta functions

Input / output functions are used to read and write data from and to external files. 
```python
read_csv()
read_excel()
DataFrame.to_csv()
DataFrame.to_excel()
read_json()
to_json()
```
Series functions are used to perform operations on series objects. 
```python
copy() # Makes a copy of an object's indices and data
iloc() # Purely integer-location based indexing for selection by position
add() # Addition of series and other, element-wise (binary operator add)
```
DataFrame functions are used to perform operations on DataFrame objects. 
Provide attributes and methods for tabluted 2D data operations.
```python
Info() # Print a concise summary of a DataFrame
values() # Return a Numpy representation of the DataFrame
size() # Return an int representing the number of elements in this object
```
Resampling functions are used to perform resampling operations on time series data. 
Plotting functions are used to plot data.
importent methods in pandas
```python
where() # Replace values where the condition is False
iterrows() # Iterate over DataFrame rows as (index, Series) pairs
head() # Return the first n rows
```
Statistical functions are used to perform some statistical operations on a data set. 
```python
mean() # Return the mean of the values for the requested axis
median() # Return the median of the values for the requested axis
mode() # Return the mode(s) of each element along the selected axis
sum() # Return the sum of the values for the requested axis
std() # Return sample standard deviation over requested axis
skew() # Return unbiased skew over requested axis
cov() # Return covariance between Series or DataFrame columns
corr() # Compute pairwise correlation of columns, excluding NA/null values
kurt() # Return unbiased kurtosis over requested axis
kutosis() # Return unbiased kurtosis over requested axis
```

window functions
```python
rolling() # Provide rolling window calculations
expanding() # Provide expanding transformations
ewm() # Provide exponential weighted functions
```

Date and Timedelta functions are used to perform some operations on date and timedelta objects. 
```python
date_range() # Return a fixed frequency DatetimeIndex
datetimeindex() # Return DatetimeIndex from an array-like of datetime-like objects
timedeltaindex() # Return TimedeltaIndex from an array-like of timedelta-like objects
to_datetime() # Convert argument to datetime
to_timedelta() # Convert argument to timedelta
timestamp() # Return POSIX timestamp as float
```
