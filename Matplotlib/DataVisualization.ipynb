Probelm Statement:

You have been provided with a dataset that list Ohio State's leading causes of death from the year 2012.

Using the two data points cause of death and percentile, draw a pie chart to visualize the dataset.

```python
# import required libraries
import matplotlib.pyplot as plt
# use down line to presnet matplot in jupyter notebook
%matplotlib inline
```

```python
# Data from Ohio state, leading causes of death in the state from year 2012
# cause of deaths
cause = 'Chronic Diseases', 'Uninterntional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others' 
# Percentile
percentile = [62, 5, 4, 2, 1, 26]
```

```python
# Set the pie chart plot properties
# Set figure size
plt.figure(figsize=(6, 6))

# Explode the largest pie in the data set
explode = (0.05, 0, 0, 0, 0, 0)

# Set pie chart properties
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)

# Set axis equal to draw pie as circle
plt.axis('equal')

# Set title of the pie chart
plt.title('Ohio State - 2012 : Leading Causes of Death')

# Show the plot (Pie Chart)
plt.show()

```
